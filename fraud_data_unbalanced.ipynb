```python
p = [fraud_data[fraud_data['Class']==i] for i in range(2)]
```

```python
train = pd.concat([p[0].iloc[0:200000, 1:], p[1].iloc[0:352, 1:]], axis = 0)
test = pd.concat([p[0].iloc[200000:, 1:], p[1].iloc[352:, 1:]], axis = 0)
```

```python
print(train.shape)
print(test.shape)
print(train.head())
print(test.head())
```

```python
from pycaret.classification import *
```
